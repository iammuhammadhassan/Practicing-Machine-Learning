```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'],style=tips['smoker'], size=tips['size'])
```

```python
sns.barplot(x=titanic['pclass'], y=titanic['fare'],hue=titanic['sex'])
```

```python
sns.boxplot(x=titanic['sex'], y=titanic['age'],hue=titanic['survived'])
```

```python
sns.kdeplot(titanic[titanic['survived']==0]['age'])
sns.kdeplot(titanic[titanic['survived']==1]['age'])
```

```python
pd.crosstab(titanic['sex'], titanic['survived'])
```

```python
pd.crosstab(titanic['pclass'], titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
titanic.groupby('pclass')['survived'].mean()
```

```python
sns.clustermap(pd.crosstab(titanic['parch'], titanic['survived']))
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.lineplot(x=flights['year'], y=flights['passengers'], hue=flights['month'])
```

```python
flights.pivot_table(index='year', columns='month', values='passengers')
```

```python
sns.heatmap(flights.pivot_table(index='year', columns='month', values='passengers'))
```

```python
sns.clustermap(flights.pivot_table(index='year', columns='month', values='passengers'))
```
