```python
Q1=df['Exam_Score'].quantile(0.25)
Q3=df['Exam_Score'].quantile(0.75)
IQR=Q3-Q1
```

```python
print("Q1:", Q1)
print("Q3:", Q3)
```

```python
upper_bound=Q3+1.5*IQR
lower_bound=Q1-1.5*IQR
```

```python
print(upper_bound)
print(lower_bound)
```

```python
df[df['Exam_Score']>upper_bound]
```

```python
df[df['Exam_Score']<lower_bound]
```

```python
new_df=df[(df['Exam_Score']<=upper_bound) & (df['Exam_Score']>=lower_bound)]
```

```python
sns.boxplot(x=new_df['Exam_Score'])
```

```python
sns.displot(new_df['Exam_Score'], bins=20,kde=True)
```

```python
new_df_cap=df.copy()
```

```python
new_df_cap = df.copy()

new_df_cap['Exam_Score'] = np.where(
    new_df_cap['Exam_Score'] > upper_bound,
    upper_bound,
    np.where(
        new_df_cap['Exam_Score'] < lower_bound,
        lower_bound,
        new_df_cap['Exam_Score']
    )
)
```

```python
sns.boxplot(x=new_df_cap['Exam_Score'])
```
